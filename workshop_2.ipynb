```python
import textstat
```

```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv(path + "/IMDB Dataset.csv")
```

```python
df['flesch_kincaid_grade'] = df['review'].apply(textstat.flesch_kincaid_grade)
display(df.head())
```

```python
df['review_cleaned'] = df['review'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
```

```python
df['review_cleaned_no_html'] = df['review_cleaned'].apply(lambda x: re.sub(r'<.*?>', '', x))
```

```python
df.head()
```

```python
df['review_clean_no extra_space'] = df['review_cleaned_no_html'].apply(lambda x: re.sub(r'\s+', ' ', x))
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
df['review_word_token'] = df['review_clean_no extra_space'].apply(lambda x: sent_tokenize(x))
```

```python
df['review_word_token_final'] =df['review_word_token'].apply(lambda x:[word_tokenize(sent) for sent in x])
```

```python
df.head()
```

```python
df['review_word_token_final_lower'] = df['review_word_token_final'].apply(lambda sentences: [[word.lower() for word in sent] for sent in sentences])
display(df.head())
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopword = stopwords.words('english')
```

```python
df['clean_stopword'] = df['review_word_token_final_lower'].apply(lambda sentences: [word for sent in sentences for word in sent if word not in stopword])
```

```python
df.head()
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
df['stemmer'] = df['clean_stopword'].apply(lambda words: [stemmer.stem(word) for word in words])
```

```python
df['review_final'] = df['stemmer'].apply(lambda words: [lemmatizer.lemmatize(word) for word in words])
```

```python
df.head()
```

```python
drop_col = ['review_cleaned','review_cleaned_no_html','review_clean_no extra_space','review_word_token','review_word_token_final','review_word_token_final_lower','clean_stopword','stemmer']
```

```python
df = df.drop(drop_col,axis=1)
```

```python
df.head()
```

```python
df['flesch_after_clean'] = df['review_final'].apply(lambda words_list: textstat.flesch_kincaid_grade(' '.join(words_list)))
```
